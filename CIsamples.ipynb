## Sample questions and solutions on Confidence Interval and Uncertainty Analysis

## Problem 1

A taxi company is trying to decide whether to purchase brand A or B tires for its fleet of taxis. An
experiment was conducted using 12 of each brand. The tires are run until they wear out. The results (in
miles) are:

\begin{array} {ccc}
\text{Brand} & \text{Sample mean} & \text{Sample standard deviation} \\
A & 36,300 & 5,000 \\
B & 38,100 & 6,100
\end{array}

* (a) Construct a 90% confidence interval for $\mu_B – \mu_A$ assuming the populations to be approximately
normally distributed. You may not assume that the variances are equal.
* (b) Does the sample information support the assumption that the variances are not equal?

## Solution
* (a) Because the variances are unknown and unequal, the T' statistic is used

\begin{equation}
(\bar{x}_1 - \bar{x}_2)-t_{\alpha/2}\sqrt{\frac{s_1^2}{n_1}+\frac{s_2^2}{n_2}} < \mu_1 - \mu_2 < (\bar{x}_1 - \bar{x}_2)+t_{\alpha/2}\sqrt{\frac{s_1^2}{n_1}+\frac{s_2^2}{n_2}}
\end{equation}
where 
\begin{equation}
\nu = \left \lfloor \frac{(s_1^2/n_1+s_2^2/n_2)^2}{(s_1^2/n_1)^2/(n_1-1) + (s_2^2/n_2)^2/(n_2-1)}\right \rfloor = 21
\end{equation}

with $t_{\alpha/2} = 1.721$ with 21 dof.

Thus the confidence interval

\begin{equation}
\mu_1 - \mu_2 = (\bar{x}_1 - \bar{x}_2) \pm t_{\alpha/2}\sqrt{\frac{s_1^2}{n_1}+\frac{s_2^2}{n_2}} = -1800 \pm 3919
\end{equation}

or $-5719 < \mu_1 - \mu_2 < 2119 ~(90\%)$

* (b) The F-distribution can be used be compare the variances of two populations

\begin{equation}
F = \frac{\sigma_B^2 S_A^2}{\sigma_A^2 S_B^2}
\end{equation}

Assuming the population variances are equal, the F-value is

\begin{equation}
f = \frac{\sigma_B^2 s_A^2}{\sigma_A^2 s_B^2} = \frac{s_A^2}{s_B^2} = 0.672
\end{equation}

Consider an f-value interval at 90% confidence level with $\nu_A = \nu_B = 11$: $P(0.355<F<2.818) = 0.90$. The f-value 0.672 falls into this range. The sample information does not support the unequal variance assumption.

Another option is to obtain the confidence interval of the ratio of variances and check if the
value of 1 falls into the interval or not. The confidence interval is

\begin{equation}
\frac{s_A^2}{s_B^2}\frac{1}{f_{\alpha/2}(\nu_1,\nu_2)} < \frac{\sigma_A^2}{\sigma_B^2} < \frac{s_A^2}{s_B^2}f_{\alpha/2}(\nu_2,\nu_1).
\end{equation}

Plug in $\frac{s_A^2}{s_B^2} = 0.672$, $f_{\alpha/2}(\nu_1,\nu_2) = f_{0.05}(11,11) = 2.818$ to get

\begin{equation}
0.672*0.355 < \frac{\sigma_A^2}{\sigma_B^2} < 0.672*2.818,
\end{equation}

which covers 1.


## Problem 2
A random sample of 20 students obtained a mean of 72 and variance of 16 on a college placement test in
mathematics. Assuming the scores to be normally distributed, construct 98% confidence intervals for $\mu$ and
$\sigma^2$, respectively.

## Solution
Given $n=20$, $\bar{x} = 72$, $s^2 = 16$
Because the population variance is unknown, the confidence interval should be based on the T statistic

$$ \mu = \bar{x} \pm t_{\alpha/2} \frac{s}{\sqrt{n}} = \bar{x} \pm t_{0.01} \frac{s}{\sqrt{n}} = 72 \pm 2.539 \frac{\sqrt{16}}{\sqrt{20}} = 72\pm 2.27 (98\%) $$

The $98\%$ confidence interval for the population variance is based on the chi-squared distribution with $\nu =
19$ and $\alpha = 0.02$

\begin{equation}
\begin{aligned}
& \frac{(n-1)s^2}{\chi_{\alpha/2}^2} < \sigma^2 < \frac{(n-1)s^2}{\chi_{1-\alpha/2}^2}\\
\rightarrow & \frac{19*16}{36.191} < \sigma^2 < \frac{19*16}{7.633} \\
\rightarrow & 8.40 < \sigma^2 < 39.83 (98\%)
\end{aligned}
\end{equation}

## Problem 3

To measure the density of a rectangular object, an experimenter measures the object's volume and mass. The volume is given by the formula $V = LWH$, where $L$ is the length, $W$ is the width, and $H$ is the height. The density $\rho$ is given by $\rho = m/V$, where $m$ is the object's mass. If the measurement of the mass is uncertain by $2\%$, and each of $L$, $W$ and $H$ is uncertain by $4\%$, what is the uncertainty, in percent, of the density $\rho$? Show your work.

## Solution

We know the uncertainty in density 

\begin{equation}
u_{\rho} = \sqrt{u_m^2 \left(\frac{\partial \rho}{\partial m}\right)^2 + u_L^2 \left(\frac{\partial \rho}{\partial L}\right)^2 + u_W^2 \left(\frac{\partial \rho}{\partial W}\right)^2 + u_H^2 \left(\frac{\partial \rho}{\partial H}\right)^2},
\end{equation}

and $u_m = 0.02m$, $u_L = 0.04L$, $u_W = 0.04W$, $u_H = 0.04H$. We also have $\frac{\partial \rho}{\partial m} = 1/V$, $\frac{\partial \rho}{\partial L} = -m/(L^2WH)$, $\frac{\partial \rho}{\partial W} = -m/(LW^2H)$, $\frac{\partial \rho}{\partial H} = -m/(LWH^2)$. Plug these in to have:

\begin{equation}
u_{\rho} = \sqrt{(0.02)^2+3*(0.04)^2}\rho
\end{equation}

## Problem 4

The experimenter conducts the same density measurement with a second sample that is spherical in shape. The mass is again uncertain by 2%. The diameter d of the sphere is measured to a precision of 4%. The volume $V$ of a sphere is given by the formula

$$ V = \frac{4\pi}{3}(\frac{d}{2})^3 $$
What is the percent uncertainty of the density ρ in this case? Show your work. Why is the uncertainty in this case different than in the case of a rectangular object? What is the underlying reason (not just how are the formulas different)?

## Solution

Similarly, we have

\begin{equation}
u_{\rho} = \sqrt{u_m^2 \left(\frac{\partial \rho}{\partial m}\right)^2 + u_d^2 \left(\frac{\partial \rho}{\partial d}\right)^2},
\end{equation}

and $u_m = 0.02m$, $u_d = 0.04d$, $\frac{\partial \rho}{\partial m} = 1/V$, and $\frac{\partial \rho}{\partial d} = -\frac{9m}{4\pi}\frac{8}{d^4}$. Plug in to get:

\begin{equation}
u_{\rho} = \sqrt{(0.02)^2+9*(0.04)^2}\rho
\end{equation}
The uncertainty is higher in this case because in all three dimensions the variances are correlated.
