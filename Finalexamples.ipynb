## Problem 1 (Bayes theorem)
A certain federal agency employs two engineering consulting firms (A, and B) with probability 0.40, and 0.60 respectively. From past experience it is known that that probabilities of cost overruns for the firms are 0.15, and 0.10, respectively.
* (a) If a cost overrun is experienced by the agency, what is the probability that it is company A?
* (b) Are the two events, "the agency has an overrun" and "company A is employed", independent? Justify your
answer.

## Solution
The chances for the two companies to be employed are $P(A) = 0.4$, $P(B) = 0.6$, with cost overrun probabilities $P(C|A) = 0.15$ and $P(C|B) = 0.1$, where $C$ represents the event of "cost overrun".

From Bayes theorem, we have $P(A|C) = \frac{P(C|A)P(A)}{P(C|A)P(A)+P(C|B)P(B)} = \frac{0.4*0.15}{0.4*0.15+0.6*0.1}$.

The event "the agency has an overrun" is $C$, while the event "company A is employed" is $A$. If these two are independent, then $P(C|A) = P(C)$ or $P(C|A) = P(C|A)P(A)+P(C|B)P(B)$. This does not hold for the given numbers.

## Problem 2 (Probability)
A student is taking MAE 318 and MAE 301 this semester. She assesses that her probability of receiving an A from MAE 318 is 0.8, and the probability that she gets an A from MAE 301 is 0.6. If, on the other hand, she believes that the probability that she will get As for both courses is 0.5, what is the probability that she will get at least one A for these two courses?

## Solution
Let "getting an A from MAE 318" be event $A$, and "getting an A from MAE 301" be event $B$. We have $P(A) = 0.8$, $P(B) = 0.6$, $P(A, B) = 0.5$. The probability of getting at least one A is $P(A or B) = P(A, \bar{B}) + P(B, \bar{A}) + P(A, B) = (P(A)-P(A,B)) + (P(B)-P(A,B)) + P(A,B) = 0.8 + 0.6 - 0.5 = 0.9$. 

## Problem 3 (t-test)
Forty force measurements are made in a lab. The sample mean is found to be $100 N$ and the sample
variance is $4 N^2$. Determine the number of additional measurements that would be required to be 80%
confident that the sample mean (of all the measurements) will be within $0.1 N$ of the true value of the force?

## Solution
According to central limit theorem, with $n = 40$, the sample mean is roughly normally distributed. Since the population variance is unknown, we have $\frac{\bar{x}-\mu}{s/\sqrt{n}}$ following a t-distribution with $n-1$ degrees of freedom. With an $80\%$ confident level, and an uncertainty limit of $0.1 N$, we have

\begin{equation}
t_{\alpha/2}(n-1)\frac{s}{\sqrt{n}} \leq 0.1.
\end{equation}

Plug in the numbers to have $n \geq \left(t_{0.1}(39)\frac{2}{0.1}\right)^2 = 680$. So, one will need another 680-40 = 640 measurements.

## Problem 4 (normal distribution, sample mean)
The life, in thousands of miles, of a certain type of electronic control for locomotives has an approximately lognormal distribution with $\mu = 5.2$ and $\sigma = 0.7$. If 50 random locomotives are selected for testing, estimate the probability that the average life of the electronic controls is greater than 5.3.

## Solution
According to central limit theorem, the sample mean (average life) follows a normal distribution with a sample size of 50. Therefore, the Z statistic $\frac{\bar{X}-\mu}{\sigma/\sqrt{n}}$ follows a standard normal distribution. The required probability of interest is

\begin{equation}
\begin{aligned}
P(\bar{X} > 5.3) & = P(z > \frac{5.3-\mu}{\sigma/\sqrt{n}})\\
& = P(z > \frac{5.3-5.2}{0.7/\sqrt{50}})\\
& = P(z > 1.01)\\
& = 1-norm.cdf(1.01)\\
& = 0.16
\end{aligned}
\end{equation}


## Problem 5 (confidence interval, uncertainty analysis)
The following are the measurements, in seconds, of the period $T$ of a vibrating system:
10.10, 9.90, 10.00
Assume the period measurement has normal distribution.
* (a) Find the 90% confidence interval for the mean spring constant AND variance of the period of the system.
* (b) If the vibrating frequency, in hertz, of the system is determined by

$$f = 1/T$$

where $T$ is the average of the period measurements, estimate the frequency of the spring with its uncertainty at confidence level $P = 90\%$. (Neglect systematic uncertainties)

## Solution
The sample mean and sample variance are $10.0$ and $0.01$, respectively. Since the population variance is unknown, the confidence interval of the mean can be calculated as $\bar{x} \pm t_{\alpha/2}(n-1) s/\sqrt{n} = 10.0 \pm 0.17$ or $[9.83, 10.17]$. The confidence interval of the variance is $[\frac{(n-1)s^2}{\chi^2_{\alpha/2}}, \frac{(n-1)s^2}{\chi^2_{1-\alpha/2}}] = [0.0033, 0.1942]$.

The uncertainty of the frequency is $u_f = |u_T \frac{d f}{d T}| = |0.17 * \frac{1}{T^2}| = |0.17/100| = 0.0017$.


## Problem 6 (confidence interval)
A hospital is considering two medications for an infection. Sample information on the length of time to recovery for patients with the problem is being used. Fifteen patients are treated with medicine 1, and seventeen patients are treated with medicine 2. The summary statistics are $\bar{x}_1 = 17$ , $s_1^2 = 1.5$, $\bar{x}_2 = 18$, and $s_2^2 = 1.8$. Assume populations are normally distributed with equal variances.
* (a) Make an appropriate assumption on the variances, and find a 99% confidence interval for the difference in the mean recovery time for the two medications.
* (b) Is your assumption on the variances reasonable? Perform statistical analysis on the assumption.

## Solution
Assume $\sigma_1^2 = \sigma_2^2$. Under this assumption, we will use a t-test. The confidence interval for $\bar{X}_1 - \bar{X}_2$ is $(\bar{x}_1-\bar{x}_2) \pm t_{\alpha/2}(n-1) \sqrt{s_p^2 (\frac{1}{n_1} + \frac{1}{n_2})}$. Plug in all numbers to get $[0.745, 3.255]$.

To test the hypothesis $\sigma_1^2 = \sigma_2^2$, we use an f-test. $f = \frac{\sigma_1^2}{\sigma_2^2} = \frac{1.5}{1.8} = 0.83$. For a confidence level of $90\%$ and $\nu_1 = 14$, $\nu_2 = 16$, we have $P(0.41<f<2.37) = 0.9$. $f=0.83$ falls into this interval. Therefore, the hypothesis cannot be rejected.

## Problem 7 (discrete random variable)
A discrete random variable $X$ has probability mass distribution
\begin{equation}
f_X(x) = \left\{ \begin{array}{cc}
0.3 & x = 3\\
0.7 & x = 5\\
0 & \text{elsewhere}
\end{array}\right.
\end{equation}

* (a) Determine the probability the value of the random variable is greater than 4.
* (b) Determine the distribution function of the sample mean $X$ (obtained with $n = 2$), and find the probability that the sample mean is greater than 4.

## Solution
$P(X>4) = P(X=5) = 0.3$

With two samples, the sample mean $\bar{X}$ has the following distribution

\begin{equation}
f_{\bar{X}}(x) = \left\{ \begin{array}{cc}
0.09 & x = 3\\
0.42 & x = 4\\
0.49 & x = 5\\
0 & \text{elsewhere}
\end{array}\right.
\end{equation}

So the sample mean is greater than 4 with probability 0.49.

$$\frac{2.1-(-8+3a)}{\sqrt{25+9a^2}} = 0.674.$$
Solve this to have $a = 1.74$.
